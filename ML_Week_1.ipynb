```python
def compute_gradient(x, y, w, b):

    m = x.shape[0]
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w * x[i] + b
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_db += dj_db_i
        dj_dw += dj_dw_i
    dj_dw = dj_dw / m
    dj_db = dj_db / m

    return dj_dw, dj_db
```

```python
def compute_cost(x, y, w, b):

    m = x.shape[0]
    cost = 0

    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i])**2
    total_cost = 1 / (2 * m) * cost

    return total_cost
```

```python
def estimate_coef(x1,y):
  beta0 = 0
  beta1 = 0
  alpha = 0.00009
  n= np.size(x1)
  y_expected = np.zeros(n)

  for j in range(100):
    for i in range(n):
      y_expected[i] = beta0 + beta1*x1[i]
      beta0 = beta0 + alpha*(y[i] - y_expected[i])*1
      beta1 = beta1 + alpha*(y[i] - y_expected[i])*x1[i]
    plt.scatter(x1,y)
    fig = plt.plot(x1,y_expected,c='orange')
    error = y-y_expected
    squared_error = np.sum(error**2)*(1/n)
    if squared_error <=1:
      print(f"squared error = {squared_error} ")
      print(f"iteration  = {n}")
  return(beta0,beta1)
```

```python
b0,b1 = estimate_coef(x1,y)

y_pred = b0+(b1*x1)
plt.plot(x1,y_pred,c='green',lw = 3)


```

```python
def estimate_coef_2(x1, y):
    beta0 = 0
    beta1 = 0
    alpha = 0.00010
    n = np.size(x1)
    y_expected = np.zeros(n)

    for j in range(100):
        for i in range(n):
            # Access values by index i instead of using the index as a key
            y_expected[i] = beta0 + beta1 * x1.iloc[i]
            beta0 = beta0 + alpha * (y.iloc[i] - y_expected[i]) * 1
            beta1 = beta1 + alpha * (y.iloc[i] - y_expected[i]) * x1.iloc[i]
        plt.scatter(x1, y)
        fig = plt.plot(x1, y_expected, c='orange')
        error = y - y_expected
        squared_error = np.sum(error**2) * (1 / n)
        if squared_error <= 1:
            print(f"squared error = {squared_error} ")
            print(f"iteration  = {n}")
    return (beta0, beta1)
```

```python

b0,b1 = estimate_coef_2(x_train,y_train)

y_pred = b0+(b1*x_train)
plt.plot(x_train,y_pred,c='blue',lw = 3)

```
