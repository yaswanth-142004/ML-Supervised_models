```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Data

```python
data = pd.read_csv("/content/sample_data/Position_Salaries.csv")
```

```python
data
```

## Splitting the Data

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
x
```

```python
y
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

## Visualize the LR Model

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title("Liear Regression")
plt.xlabel("Position Level")
plt.ylabel("Prediction")
plt.show()
```

## Polynomial Linear regression

```python
from sklearn.preprocessing  import PolynomialFeatures

for i in range(10):


  poly_reg = PolynomialFeatures(degree = i)

  x_poly = poly_reg.fit_transform(x)

  poly_linear_reg = LinearRegression()

  poly_linear_reg.fit(x_poly,y)



  plt.scatter(x,y,color='red')
  plt.plot(x,poly_linear_reg.predict(poly_reg.fit_transform(x)),color='blue')
  plt.title(f"Polynomial Regression  for degree {i}" )
  plt.xlabel("Position Level")
  plt.ylabel("Prediction")
  plt.show()

```
